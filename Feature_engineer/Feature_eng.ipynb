Creo una nueva columna llamada 'análisis_sentimiento' reemplazando la columna 'reviews'. 
Realizp un análisis de sentimiento  con los parametros que se nos entregaron 

0 si es negativo
1 si es neutral o sin reseña
2 si es positivo

```python
""" Aplico la característica de análisis de sentimiento"""

df_user_reviews['análisis_sentimiento'] = df_user_reviews['review'].apply(lambda x: TextBlob(str(x)).sentiment.polarity if pd.notna(x) and x != 'Sin datos' else None)

def mapear_sentimiento(puntuación):
    if puntuación is None:
        return 1  # Si la reseña está ausente o es 'Sin datos', toma el valor de 1
    elif puntuación < 0:
        return 0  # malo
    elif puntuación == 0:
        return 1  # neutral
    else:
        return 2  # positivo

df_user_reviews['análisis_sentimiento'] = df_user_reviews['análisis_sentimiento'].apply(mapear_sentimiento)
```

PLaytime Genre
creo un nuevo dataframe que muestre la relacion entre las columnas 'genres', "year' 
y horas jugadas por genero

```python
"""las columnas 'playtime_forever' y 'item_id' se extraen del dataframe 'df_users_items' """ 

playtime_Uitem = df_users_items[['playtime_forever', 'item_id']]
playtime_Uitem
```

```python
"""las columnas 'item_id', 'genres' y 'date' las extraigo del dataframe 'df_steam_games'"""

df_genre_year = df_steam_games[['item_id', 'genres', 'date']]
```

```python
df_genre_year
```

```python
"""uno los dos dataframes resultantes para poder obetenrr el dataframe resultante 'df_playtime_genre' y asi poder 
utilizar este en la funcion """

df_playtime_genre = df_genre_year.merge(playtime_Uitem, on='item_id')
df_playtime_genre
```

```python
""" ahora agrupo las filas por genero y year, lo cual me da como resultado
la suma del tiempo de juego """ 

playtime_genre = df_playtime_genre.groupby(['genres', 'date'])['playtime_forever'].sum().reset_index()
playtime_genre
```

User for Genre

ahora creamos  un DataFrame a que nos muestre 'genres', 'year', 'user' y las horas 
jugadas de cada  usuario por año en cada género.



```python
users_items = df_users_items[['playtime_forever', 'user_id', 'item_id']]
users_items.head()
```

```python
user_for_genre=df_genre_year.merge(users_items, on='item_id')
user_for_genre
```

```python
"""convierto el timepo en minutos a horas """

user_for_genre['playtime_hours'] = user_for_genre['playtime_forever'] / 60
user_for_genre['playtime_hours'] = user_for_genre['playtime_hours'].round(2)
```

```python
user_for_genre = user_for_genre.groupby(['genres', 'user_id', 'date'])['playtime_hours'].sum().reset_index()
user_for_genre
```

```python
"""Para ahorrar memoria en las funciones y modelos futuros, tomamos una muestra aleatoria de datos. """

random_rows = 2000000
user_for_genre = user_for_genre.sample(n=random_rows, random_state=42)
user_for_genre
```

```python
user_for_genre = user_for_genre.groupby(['genres', 'user_id', 'date'])['playtime_hours'].sum().reset_index()
user_for_genre
```
