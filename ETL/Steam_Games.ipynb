ATRAVES DE LA FUNCION ".INFO ( ) " PODEMOS OBSERVAR QUE HAY UNA GRAN CANTIDAD DE DATOS NULOS O  FALTANTES POR LO CUAL CALCULAREMOS LA CANTIDAD DE ESTOS Y SUS PORCENTAJES


```python
def resumen_valores_faltantes(df: pd.DataFrame) -> pd.DataFrame:
  """
   Esta funcion analiza la cantidad y el porcentaje de valores faltantes (NaN) en cada columna de un DataFrame.

  ParÃ¡metros:
    df: El DataFrame a analizar.

  Devuelve:
    Un DataFrame que resume la informaciÃ³n de valores faltantes:
      - #missing: Cantidad de valores faltantes en cada columna.
      - %missing: Porcentaje de valores faltantes en cada columna.
  """

  # ValidaciÃ³n del tipo de dato
  if not isinstance(df, pd.DataFrame):
    raise TypeError("El argumento 'df' debe ser un DataFrame de Pandas.")

  # CÃ¡lculo de resumen
  resumen = pd.DataFrame(
      index=df.columns,
      columns=["#missing", "%missing"],
      dtype=int,
  )
  resumen["#missing"] = df.isnull().sum()
  resumen["%missing"] = (resumen["#missing"] / len(df)) * 100

  # Formateo de la columna "%missing"
  resumen["%missing"] = resumen["%missing"].map("{:.1f}%".format)

  return resumen

# Ejemplo de uso con el nombre del DataFrame
resultado = resumen_valores_faltantes(Ruta_Steam_Games)
print(resultado.to_string())
```

Como podemos observar este dataframe continen mas del 70% de sus datos nulos por lo cual procedere a eliminarlos, ya que esto no afectara  la calidad de los resultados para los endpoints requeridos 


```python
def eliminar_filas_con_nulos(df):
  """
  Elimina las filas que contienen valores nulos en un DataFrame.

  ParÃ¡metros:
    df: El DataFrame a analizar.

  Devuelve:
    Un DataFrame con las filas que contienen valores nulos eliminadas.
  """

  # Filtra las filas donde no hay valores nulos en ninguna columna
  df_sin_nulos = df.dropna(axis=0)

  # Imprime la cantidad de filas eliminadas
  filas_eliminadas = df.shape[0] - df_sin_nulos.shape[0]
  print(f"Se eliminaron {filas_eliminadas} filas con valores nulos.")

  return df_sin_nulos

# Ejemplo de uso
df_games_sin_nulos = eliminar_filas_con_nulos(Ruta_Steam_Games.copy())
```

```python
""" Puedo ovseervar que los valores de la columna price no se relacionan , por lo cual intentare normalizarlos a dos decimales """
# Convierto valores no numÃ©ricos a 0 y redondear a 2 decimales
def convertir_a_numero(valor):
    try:
        # Intenta convertir el valor a un nÃºmero y redondearlo a 2 decimales
        return round(float(valor), 2)
    except (ValueError, TypeError):
        # Si no se puede convertir, devuelve 0
        return 0

# Aplica la funciÃ³n a la columna 'price'
Df_steam_games['price'] = Df_steam_games['price'].apply(convertir_a_numero)
Df_steam_games
```
