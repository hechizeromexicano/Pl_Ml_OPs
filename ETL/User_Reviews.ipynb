```python
"""Primero cargo las librerias necesarias para poder trabajar adecuadamente con los datos que
contiene este archivo """

import pandas as pd
import numpy as np 
import ast 
import json
import gzip 
import re

```

```python
""" defino la ruta, cargo el archivo a analizar y lo transformo en un dataframe """

Ruta_user_reviews = []
with gzip.open('C:/Users/Daniel/Downloads/PI_01_07/Data/user_reviews.json.gz', 'rb') as f : 
    for line in f:
        Ruta_user_reviews.append(ast.literal_eval(line.decode('utf_8')))
Df_user_reviews = pd.DataFrame(Ruta_user_reviews)
f.close()
```

observo que la columna reviews posee valores anidados por lo cual procedere a desanidar 


```python
reviews_desanidados = []

# Itero a traves de cada fila del df
for index, row in Df_user_reviews.iterrows():
    # Obtengo los valores de las columnas
    user_id = row['user_id']
    user_url = row['user_url']
    for review_dict in row['reviews']:
        review_dict['user_id'] = user_id
        review_dict['user_url'] = user_url
        reviews_desanidados.append(review_dict)

# Actualizo el DataFrame con la lista de datos desanidados 
        
Df_user_reviews = pd.DataFrame(reviews_desanidados)
```

```python
"""Reemplazo valores vacios por None """

Df_user_reviews.replace('', None, inplace=True)
Df_user_reviews
```

```python
""" Cambio el formato de la fecha, ya que solo necesito el dato 'year' """

def date_post(date_str):
    date_r = re.search(r'(\w+\s\d{1,2},\s\d{4})', date_str)
    if date_r:
        date_rvw = date_r.group(1)
        try:
            date_rt = pd.to_datetime(date_rvw)
            return date_rt.strftime('%Y')
        except ValueError:
            return 'Invalid Date'
    else:
        return 'Invalid Format'

Df_user_reviews['posted'] = Df_user_reviews['posted'].apply(date_post)

Df_user_reviews
```

```python
Df_user_reviews = Df_user_reviews[Df_user_reviews['posted'] != 'Invalid Format']
Df_user_reviews['posted'] 
```

```python
""" Ahora elimino las columnas que no voy a utilizar """

Df_user_reviews.drop(['funny', 'last_edited', 'helpful', 'user_url'], axis=1, inplace=True)

Df_user_reviews
```
