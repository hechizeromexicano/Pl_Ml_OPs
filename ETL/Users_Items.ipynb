```python
"""Cargamos las librerias necesarias para poder trabajar este archivo adecuadamente"""
import pandas as pd  
import ast 
import re
import gzip
import json
import os  
```

```python

Ruta_users_items = []

with gzip.open('C:/Users/Daniel/Downloads/PI_01_07/Data/users_items.json.gz', 'rb') as f : 
    for line in f:
        Ruta_users_items.append(ast.literal_eval(line.decode('utf_8')))
Df_users_items = pd.DataFrame(Ruta_users_items)
f.close()
```

```python
""" puedo observar que este dataset  contiene  una columna con datos anidados, pore lo cual, procedo a desanidarlo"""

# Lista vacÃ­a para almacenar los datos desanidados
items_desanidados = []

# Iterar a travÃ©s de cada fila del DataFrame df_items
for indice, row in Df_users_items.iterrows():
    # Obtener los valores de las columnas de interÃ©s de cada fila
    user_id = row['user_id']
    steam_id = row['steam_id']
    user_url = row['user_url']

    # Iterar a travÃ©s de los elementos de la columna 'items' de la fila actual
    for i in row['items']:
        # Crear un nuevo diccionario con los datos desanidados de cada Ã­tem
        nuevo = {
            'user_id': user_id,
            'steam_id': steam_id,
            'user_url': user_url,
            'item_id': i['item_id'],
            'item_name': i['item_name'],
            'playtime_forever': i['playtime_forever'],
            'playtime_2weeks': i['playtime_2weeks']
        }
        # Agregar el diccionario a la lista 'desanidado'
        items_desanidados.append(nuevo)

# Actualizo df_items a partir de la lista de datos desanidados
Df_users_items = pd.DataFrame(items_desanidados)
```

```python
# Imprimir encabezado
print("InformaciÃ³n de tipos de datos y valores nulos:")

# Obtener tipos de datos
data_types = Df_users_items.dtypes

# Calcular porcentaje de valores nulos
missing_values_percent = (Df_users_items.isnull().sum() / len(Df_users_items)) * 100

# Crear DataFrame
df_info = pd.DataFrame({
  "nombre_columna": data_types.index,
  "tipo_datos": data_types.values,
  "no_nulos": Df_users_items.shape[0] - Df_users_items.isnull().sum(axis=0),
  "%nulos": missing_values_percent.round(2)
})

df_info
```
